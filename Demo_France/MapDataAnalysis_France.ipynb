```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import json
with open('numMapData.json','r')as f:
    data = json.load(f)
    print(data)
```

```python
dataFrame = pd.DataFrame(data['data'])
dataFrame
```

```python
rows, cols = 6, 4
sumIndex = 0 
fig,ax = plt.subplots(rows,cols,figsize=(10,20),sharex=True, sharey=True)
for i in range(rows):
    for j in range(cols):
        a = dataFrame.iloc[sumIndex]
        b = [i for i in range(len(a))]
        ax[i][j].plot(b,a)
        sumIndex = sumIndex+1
```

```python
rows, cols = 6, 4
sumIndex = 0 
fig,ax = plt.subplots(rows,cols,figsize=(10,20),sharex=True, sharey=True)
for i in range(rows):
    for j in range(cols):
        a = dataFrame.iloc[sumIndex]
        b = [i for i in range(len(a))]
        ax[i][j].hist(a,bins=20)
        sumIndex = sumIndex+1
```

```python
rows, cols = 6, 4
sumIndex = 0 
fig,ax = plt.subplots(rows,cols,figsize=(10,20),sharex=True, sharey=True)
for i in range(rows):
    for j in range(cols):
        a = dataFrame.iloc[sumIndex]
        b = [i for i in range(len(a))]
        ax[i][j].hist(a,bins=[0,0.25,0.5,0.75,1])
        sumIndex = sumIndex+1
```
